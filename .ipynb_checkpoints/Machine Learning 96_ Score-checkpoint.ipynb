```python
#Machine Learning

X = df.drop('Price',axis=1)
y = df['Price']
```

```python
numeric_features = X.select_dtypes(include=np.number).columns
catergoric_features = X.select_dtypes(include=object).columns
```

```python
numeric_features
```

```python
catergoric_features
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

label_enc = LabelEncoder()

for col in catergoric_features:
    X[col] = label_enc.fit_transform(X[col])
    
scaler = StandardScaler(with_mean=True, with_std=True)
X = scaler.fit_transform(X)

scaler = StandardScaler(with_mean=True, with_std=True)
y = scaler.fit_transform(pd.DataFrame(y))
```

```python
import catboost as cat
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=1/3,shuffle=True,random_state=1001)


model = cat.CatBoostRegressor(verbose=False,allow_writing_files=False)

model.fit(x_train,y_train.ravel()).score(x_test,y_test)
```
